##  Veri Setinde Sınıflandırma İşlemleri

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("heart_failure_clinical_records_dataset.csv")
print(df.info())
```

```python
y=df.DEATH_EVENT.values
x_data=df.drop(["DEATH_EVENT"],axis=1)
```

```python
x=x_data
#x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

### KNN Algoritması

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
knn_score=knn.score(x_test,y_test)
print(knn_score)
```

### SVM ALgoritması

```python
from sklearn.svm import SVC
svm=SVC(random_state=1)
svm.fit(x_train,y_train)
svm_pred=svm.predict(x_test)
print("svm score:",svm.score(x_test,y_test))
```

### Decision Tree Algoritması

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=1)
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
print("dt score:",dt.score(x_test,y_test))
```

### Random Forest Algoritması

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=200, random_state=1)
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
print("rf score:",rf.score(x_test,y_test))
```

```python
# Bu veri setinde normalızasyon yapılmış ve normalizasyon yapılmamış datalar sınıflandırma algoritmaları çalıştırıldığında aynı sonuçları döndürdüğü görülmüştür.
```

##  Sınıflandırma Algoritmalarının Değerlendirilmesi

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

### KNN İçin Değerlendirme 

```python
y_pred= pd.Series.from_array(knn_pred)
y_test2 = pd.Series.from_array(y_test)
metrics.confusion_matrix(y_test2, y_pred)
```

```python
print("Confusion Matris İle Değerlendirme")
print("")
print(" Doğruluk Oranı : ","Doğruluk = TN + TP / TOPLAM")
dogruluk = metrics.accuracy_score(y_test2, y_pred)
print(dogruluk)
print(" ")
print("Hata Oranı : ","Hata Oranı = FN + FP / TOPLAM veya 1 - dogruluk")
Hata_Oranı=1-dogruluk
print(Hata_Oranı)
print(" ")
print("Precision : ","Precision = TP/(FP+TP) idi. Yani pozitif tahminlerin toplam pozitiflere oranı.")
metrics.precision_score(y_test2, y_pred)
```

```python
print("KNN ")

print(classification_report(y_test2, y_pred))
```
